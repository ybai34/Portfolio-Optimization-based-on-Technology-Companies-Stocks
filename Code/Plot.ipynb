```python
import pandas as pd
import numpy as np

# Data provided
data = {
    "default": (0.001831, 0.019400, 0.094378, 0.669910),
    "1.4": (0.002563, 0.019400, 0.094378, 0.669910),
    "1.6": (0.002930, 0.030679, 0.095492, 0.744083),
    "1.8": (0.003296, 0.037693, 0.087436, 0.827500),
    "comp_mean": (0.0004586638988146737, 0.016919727216098274, 0.02710823247660151, 0.6730260279613982)
}

# Creating a DataFrame for easier manipulation
df = pd.DataFrame(data, index=["Expected Return", "Risk", "Sharpe Ratio", "Diversification"]).T

# Displaying the DataFrame for verification
df
```

```python
from math import pi

# Function to plot a radar chart for a given data point
def plot_radar_chart(df, labels, title):
    # Number of variables
    categories = list(df)
    N = len(categories)

    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Initialise the spider plot
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)

    # Draw one axe per variable + add labels
    plt.xticks(angles[:-1], categories)

    # Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.5", "0.75"], color="grey", size=7)
    plt.ylim(0, 1)

    # Plot each individual = each line of the data
    for label in labels:
        values = df.loc[label].values.flatten().tolist()
        values += values[:1]
        ax.plot(angles, values, linewidth=1.5, linestyle='solid', label=label)
        ax.fill(angles, values, alpha=0.1)

    # Add legend
    plt.title(title, size=20, y=1.1)
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

    plt.show()

# Adjusting the DataFrame for radar chart (inverting Diversification as lower is better)
df_radar = df.copy()
df_radar["Diversification"] = 1 - df_radar["Diversification"] # Invert the scale for Diversification
df_radar = df_radar / df_radar.max() # Normalize the data

# Plotting the radar chart for all data points
plot_radar_chart(df_radar, df_radar.index, "Comparison of Investment Models")
```

```python
# Function to plot a radar chart for a single data point
def plot_single_radar_chart(df, label, title):
    categories = list(df)
    N = len(categories)

    values = df.loc[label].values.flatten().tolist()
    values += values[:1]  # Completing the loop

    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], categories)
    ax.plot(angles, values)
    ax.fill(angles, values, 'teal', alpha=0.1)

    plt.title(title, size=20, y=1.1)
    plt.show()

# Plotting radar chart for the first data point ("default")
plot_single_radar_chart(df_radar, "default", "Radar Chart for 'default' Data Point")
```
